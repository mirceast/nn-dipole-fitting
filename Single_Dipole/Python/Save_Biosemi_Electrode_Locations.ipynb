I generate electrode locations of Biosemi montages and save them in the `data_dir` folder. I then use them in Matlab to generate separate forward structures. For some reason I have to multiply them by 10000. 

```python
import mne
import numpy as np
%matplotlib inline 
%config InlineBackend.figure_format = 'retina'

# Parameters for loading the .mat file
data_dir = "C:/Users/Mircea/Google Drive/Data/nn-dipole-fitting/"

layouts = [16,32,64,128,256]
for layout in layouts:
    montage = mne.channels.read_montage(f"biosemi{layout}",transform=False)
    np.savetxt(f"{data_dir}Biosemi_{layout:03d}.csv", np.array(montage.pos), delimiter=",")
```
