```python
# Center of the brain for analyzing error vs depth
brain_center = np.array([0.0323, -1.6413,0.2302])

# Load predicted data
dipoles_nn = load_dict("dipoles_nn.pickle")
dipoles_lm = load_dict("dipoles_lm.pickle")
n_snr = len(dipoles_nn)

list(dipoles_nn[1])
```

```python
from scipy.stats import wasserstein_distance
import time

x = np.random.randn(2113)
y = np.random.randn(2113)
t = time.time()
wasserstein_distance(x,y)
print((time.time() - t)*1e6)

from pyemd import emd, emd_samples
t = time.time()
emd_samples(x, y)
print((time.time() - t)*1e6)
```

```python
from misc_analysis import *

for i in range(len(dipoles_nn)):
    this_nn = dipoles_nn[i]
    this_lm = dipoles_lm[i]

    hist_centers, hist_nn, hist_lm = get_err_hist(dipoles_nn[i], dipoles_lm[i])

    # Find out the distance from the brain center to all predicted dipoles
    dist = np.sqrt(((this_nn["loc"] - brain_center)**2).sum(axis=1))
    # Now assume the maximum distance is to the brain surface and convert to depth
    depth = np.abs(dist - dist.max())
    # Divide the depth values in a few steps    
    depth_centers, depth_edges = get_hist_edges(data=depth, n_bins=5)
    # Find the data the falls in each category
    means_nn, stds_nn = get_means_and_stds_in_hist_edges(depth_edges, depth, this_nn["loc_err"])
    means_lm, stds_lm = get_means_and_stds_in_hist_edges(depth_edges, depth, this_lm["loc_err"])

    fig, ax = plt.subplots(1,2,figsize=(12,4))
    ax[0].plot(hist_centers, hist_nn, label='NN')
    ax[0].plot(hist_centers, hist_lm, label='LM')
    ax[0].legend()
    ax[0].set_xlabel("Error [cm]")
    ax[0].set_ylabel("Count")    
    ax[0].set_title("Histograms of localization error\nMean NN = {:.3f}, LM = {:.3f} cm".format(
                    np.mean(this_nn["loc_err"]), np.mean(this_lm["loc_err"])))

    ax[1].errorbar(depth_centers, means_nn, yerr=stds_nn, label='NN', marker='o')
    ax[1].errorbar(depth_centers, means_lm, yerr=stds_lm, label='LM', marker='o')
    ax[1].set_xlabel("Depth [cm]")
    ax[1].set_ylabel("Error [cm]")
    ax[1].legend()
    ax[1].set_title("Localization error vs source depth, M  SD")
    fig.suptitle("SNR = {:.2f}".format(this_nn["snr"]), y=1.05, fontsize=16)
    plt.show()

```
